```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ncr=pd.read_csv("ncr_ride_bookings.csv")
ncr.head()
```

```python
##1)Convert Date and Time to datetime

pd.to_datetime(ncr['Date'])
pd.to_datetime(ncr['Time'])
```

**2)Combine Date and Time into a single datetime column**

```python
##2)Combine Date and Time into a single datetime column

ncr['datetime']=pd.to_datetime(ncr['Date'] + ' ' + ncr['Time'])
ncr.head()
```

```python
##3)Extract temporal features**

ncr['year']=ncr['datetime'].dt.year
ncr['month']=ncr['datetime'].dt.month
ncr['day']=ncr['datetime'].dt.day
ncr['week']=ncr['datetime'].dt.day_name()
ncr['hour']=ncr['datetime'].dt.hour
ncr['minutes']=ncr['datetime'].dt.minute
ncr['second']=ncr['datetime'].dt.second
ncr.head()
```

```python
##4)Create binary flags from cancellation/incomplete columns

ncr['flag_customer_cancel']=ncr['Cancelled Rides by Customer'].notna().astype(int)
ncr['flag_driver_cancel']=ncr['Cancelled Rides by Driver'].notna().astype(int)
ncr['flag_incomplete_ride']=ncr['Incomplete Rides'].notna().astype(int)
ncr.head()
```

```python
##5)Create a flag for missing ratings and booking values

ncr['flag_missimg_ratings_driver']=ncr['Driver Ratings'].isna().astype(int)
ncr['flag_missing_rating_customer']=ncr['Customer Rating'].isna().astype(int)
ncr['flag_booking_value']=ncr['Booking Value'].isna().astype(int)
ncr.head()
```

```python
##6)Define Target for Customer Cancellation

ncr['flag_customer_cancel'].value_counts()
```

```python
##7)Cancellation Rate by Vehicle Type

cancel_counts = ncr.groupby('Vehicle Type')['flag_customer_cancel'].sum()
plt.pie(cancel_counts.values, labels=cancel_counts.index, autopct='%1.1f%%')
plt.title("Customer Cancellation by Vehicle Type (Counts)")
plt.show()
```

```python
##8)Cancellation Rate by Hour

ncr['hour']=ncr['datetime'].dt.hour
cancel_rate_hour=(ncr.groupby('hour')['flag_customer_cancel'].mean())*100
plt.plot(cancel_rate_hour.values,cancel_rate_hour.index,marker='*')
```
