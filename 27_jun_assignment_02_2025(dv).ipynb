# Function Summary:-

```python
def batsman_info(dv,batsman_name):
    bm_ipl=dv[dv["batsman"]==batsman_name]
    total_runs=bm_ipl["batsman_runs"].sum()
    total_balls=bm_ipl.shape[0]
    four=bm_ipl[bm_ipl["batsman_runs"]==4].shape[0]
    six=bm_ipl[bm_ipl["batsman_runs"]==6].shape[0]
    matches=bm_ipl["match_id"].nunique()
    strike_rate = (total_runs / total_balls) * 100
    
    
    info={
          "batsman_name":batsman_name,
          "total_runs":total_runs,
          "balls":total_balls,
          "Fours": four,
          "Sixes": six,
          "Matches": matches,
          "Strike Rate": round(strike_rate, 2)
        
    }
    
    return pd.DataFrame([info])                           
```

```python
batsman_info(dv,"DA Warner")
```
