```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
df.groupby('sex')[['total_bill']].sum()
```

```python
df.groupby(['sex','smoker'])[['total_bill']].mean()
```

```python
df.pivot_table(index='sex',columns='smoker',values='total_bill')
```

```python
df.pivot_table(index='sex',columns='smoker')
```

```python
df.pivot_table(index='sex',columns='smoker',aggfunc='sum')
```

```python
df.pivot_table(index='sex',columns='smoker',values='total_bill',aggfunc='sum')
```

```python
df.pivot_table(index=['sex','smoker'],columns=['day','time'])
```

```python
df['day'].unique()
```

```python
df.pivot_table(index=['sex','smoker'],columns=['day','time'],values='total_bill')
```

```python
df.pivot_table(index=['sex','smoker'],columns=['day','time'],values='total_bill',aggfunc='sum')
```

```python
df.pivot_table(index=['sex','smoker'],columns=['day','time'],aggfunc={'size':'mean','tip':'max','total_bill':'sum'})
```

```python
df.pivot_table(index=['sex','smoker'],columns=['day','time'],aggfunc={'size':'mean','tip':'max','total_bill':'sum'},margins=True)
```

```python
df.pivot_table(index=['sex','smoker'],columns=['day','time'],values='total_bill',aggfunc='sum',margins=True)
```
