```python
import numpy as np
import pandas as pd
```

```python
course = pd.read_excel("course_details.xlsx")
course.head()
```

```python
student = pd.read_excel("student_details.xlsx")
student.head()
```

```python
May = pd.read_excel("May.xlsx")
May.head()
```

```python
Jun = pd.read_excel("Jun.xlsx")
Jun.head()
```

```python
May_Jun = pd.concat([May,Jun])
May_Jun
```

```python
May_Jun = pd.concat([May,Jun],ignore_index=True)
May_Jun
```

```python
May.append(Jun,ignore_index=True)
```

```python
both = pd.concat([May,Jun],keys=['May','Jun'])
both
```

```python
both.loc['May']
```

# Inner Join:-

```python
student.merge(May_Jun,how='inner',on='sid')
```

# Left Join:-

```python
student.merge(May_Jun,how='left',on='sid')
```

# Right Join:-

```python
student.merge(May_Jun,how='right',on='sid')
```

# Full Outer Join:-

```python
student.merge(May_Jun,how='outer',on='sid')
```
